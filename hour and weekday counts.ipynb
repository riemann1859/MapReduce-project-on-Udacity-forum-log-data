```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
add='C:/Users/YUNUS/Desktop/part-00000'
result=pd.read_csv(add,header=None,sep='\t')
result.columns=['type','label','count']
hour=result[result['type']=='hour']
hour.columns=['type','hour','count']
hour.sort_values(by='hour',inplace=True)
weekday=result[result['type']=='weekday']
weekday.columns=['type','weekday','count']
weekday=weekday.sort_values(by='weekday').iloc[:7,1:3]
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=hour['hour'],y=hour['count'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],y=weekday['count'])
plt.xticks(rotation=45)
```
